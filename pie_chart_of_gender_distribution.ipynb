### Pie Chart of Gender Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('Figure Data.xlsx','Sheet1')
labels=['Male','Female']
colors=[(125/255,167/255,217/255),(199/255,69/255,66/255)]

plt.figure(figsize=(10,4))

plt.subplot(121)
plt.title('06106',fontsize=14)
plt.pie(df.iloc[0,1:],labels=labels,colors=colors,counterclock=False,startangle=90,autopct='%0.1f%%')
plt.subplot(122)
plt.title('06112',fontsize=14)
plt.pie(df.iloc[1,1:],labels=labels,colors=colors,counterclock=False,startangle=90,autopct='%0.1f%%')
plt.figlegend(labels=['Male','Female'],loc='upper center')

plt.savefig('gender.svg', dpi=600, bbox_inches='tight')
plt.show()
```
