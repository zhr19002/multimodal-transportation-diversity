### Bar Chart of Race Distribution (Vertical)

```python
import pandas as pd
import matplotlib.pyplot as plt

df3=pd.read_excel('Figure Data.xlsx','Sheet3')
zipcodearea=['06106','06112','06114','06120']
width=0.4
interval=0.3
x1=[]
x2=[]
x3=[]
x4=[]
x5=[]
for i in range(len(zipcodearea)):
    temp=(i+1)*interval+((df3.shape[1]-1)*i+0.5)*width
    x1.append(temp)
    x2.append(temp+width)
    x3.append(temp+2*width)
    x4.append(temp+3*width)
    x5.append(temp+4*width)

plt.figure(figsize=(10,6))
plt.bar(x1, df3['White'], width=width, label='White', color=(237/255,125/255,49/255))
plt.bar(x2, df3['Hispanic'], width=width, label='Hispanic', color=(68/255,114/255,196/255))
plt.bar(x3, df3['Asian'], width=width, label='Asian', color=(255/255,192/255,0/255))
plt.bar(x4, df3['Black'], width=width, label='Black', color=(165/255,165/255,165/255))
plt.bar(x5, df3['Other'], width=width, label='Other', color=(91/255,155/255,213/255))
plt.xlabel('Zip Code')
plt.ylabel('Percentage')
plt.legend(loc='upper right')
plt.xticks(x3,zipcodearea)

for a,b in zip(x1,df3['White']):
    plt.text(a,b+0.005,'{:.1%}'.format(b),ha='center', va='bottom')
for a,b in zip(x2,df3['Hispanic']):
    plt.text(a,b+0.005,'{:.1%}'.format(b),ha='center', va='bottom')
for a,b in zip(x3,df3['Asian']):
    plt.text(a,b+0.005,'{:.1%}'.format(b),ha='center', va='bottom')
for a,b in zip(x4,df3['Black']):
    plt.text(a,b+0.005,'{:.1%}'.format(b),ha='center', va='bottom')
for a,b in zip(x5,df3['Other']):
    plt.text(a,b+0.005,'{:.1%}'.format(b),ha='center', va='bottom')

plt.savefig('race.svg', dpi=600, bbox_inches='tight')
plt.show()
```
