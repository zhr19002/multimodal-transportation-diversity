### Pie Chart of Age Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Figure.xlsx','Sheet2')
labels = ['Under 15','15-24','25 and above']
colors = [(125/255,167/255,217/255),(199/255,69/255,66/255),(255/255,236/255,72/255)]

plt.figure(figsize = (10,4))

plt.subplot(121)
plt.title('06106', fontsize=14)
plt.pie(df.iloc[0,1:], labels=labels, colors=colors, counterclock=False, startangle=90, autopct='%0.1f%%')
plt.subplot(122)
plt.title('06112', fontsize=14)
plt.pie(df.iloc[1,1:], labels=labels, colors=colors, counterclock=False, startangle=90, autopct='%0.1f%%')
plt.figlegend(labels=labels, loc='upper center')

#plt.savefig('Age.svg', dpi=600, bbox_inches='tight')
plt.show()
```
