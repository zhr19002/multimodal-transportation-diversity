### Line Chart of Diversity Accross Zip Code Areas

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('Figure Data.xlsx','Sheet4')
x=[1,2,3,4,5,6]

plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.plot(x, df['Physical Infrastructure Diversity'], 's-', label='Physical Infrastructure Diversity',color=(68/255,114/255,196/255))
for a,b in zip(x,df['Physical Infrastructure Diversity']):
    plt.text(a,b+0.005,'{:.3}'.format(b),ha='center', va='bottom')

plt.xlabel('Zip Code')
plt.ylabel('Diversity Values')
plt.legend(loc='best')
plt.xticks(x,['06103','06105','06106','06120','06112','06114'])

plt.subplot(2,1,2)
plt.plot(x, df['Travel Behavior Diversity'], 'o-', label='Travel Behavior Diversity',color=(237/255,125/255,49/255))
for a,b in zip(x,df['Travel Behavior Diversity']):
    plt.text(a,b+0.005,'{:.3}'.format(b),ha='center', va='bottom')

plt.xlabel('Zip Code')
plt.ylabel('Diversity Values')
plt.legend(loc='best')
plt.xticks(x,['06103','06105','06112','06120','06106','06114'])

plt.savefig('diversity.svg', dpi=600, bbox_inches='tight')
plt.show()
```
