### SVI-SDI-RI Dataset

```python
import numpy as np
import pandas as pd

#Social vulnerability features
SV = ['EPL_POV150','EPL_UNEMP','EPL_HBURD','EPL_NOHSDP','EPL_UNINSUR','RPL_THEME1',
      'EPL_AGE65','EPL_AGE17','EPL_DISABL','EPL_SNGPNT','EPL_LIMENG','RPL_THEME2',
      'EPL_MINRTY',
      'EPL_MUNIT','EPL_MOBILE','EPL_CROWD','EPL_NOVEH','EPL_GROUPQ','RPL_THEME4']

SVI = pd.read_csv('SVI_Dataset.csv')
df = pd.read_csv('SDI_RI_Dataset.csv')

for item in SV:
    df[item] = -999
    ZIP_SV_dic = dict(zip(SVI['NAME'], SVI[item]))
    for key,value in ZIP_SV_dic.items():
        if key in list(df['NAME']):
            df.loc[df[df['NAME'] == key].index.tolist()[0], item] = value

#Drop rows
df = df.replace(-999, np.nan)
df = df.dropna()

#Data output
df.to_csv('SVI_SDI_RI_Dataset.csv',index=False)
```
